```python
import os
import sys
import platform

# 1. Інформація про ОС
os_info = f"{platform.system()} {platform.release()} ({platform.machine()})"

# 2. Шлях до папки проекту
project_path = os.getcwd()

# 3. Назва віртуального оточення
# Перевіряємо шлях до виконуваного файлу Python
venv_path = sys.prefix
venv_name = os.path.basename(venv_path)

print(f"Операційна система: {os_info}")
print(f"Шлях до проекту: {project_path}")
print(f"Віртуальне оточення: {venv_name}")
```
