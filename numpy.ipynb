## Завдання 1

$$z_1 = \cos^2 \left( \frac{3}{8}\pi - \frac{\alpha}{4} \right) - \cos^2 \left( \frac{11}{8}\pi + \frac{\alpha}{4} \right)$$
$$z_2 = \frac{\sqrt{2}}{2} \sin \frac{\alpha}{2}$$

```python
import numpy as np

alpha = np.pi / 4

# 2. Обчислення z1
# Формула: cos^2(3/8*pi - alpha/4) - cos^2(11/8*pi + alpha/4)
term1 = np.cos(3/8 * np.pi - alpha/4)**2
term2 = np.cos(11/8 * np.pi + alpha/4)**2
z1 = term1 - term2

# 3. Обчислення z2
# Формула: (sqrt(2)/2) * sin(alpha/2)
z2 = (np.sqrt(2) / 2) * np.sin(alpha / 2)

# 4. Вивід результатів
print(f"Параметр alpha: {alpha}")
print(f"Результат z1: {z1}")
print(f"Результат z2: {z2}")

# Додаткова перевірка на тотожність (різниця має бути близькою до нуля)
print(f"Різниця між z1 та z2: {abs(z1 - z2)}")
```

## Завдання 2


$$\text{arcctg } x = \frac{\pi}{2} + \sum_{n=0}^{\infty} \frac{(-1)^{n+1} x^{2n+1}}{2n+1} = \frac{\pi}{2} - x + \frac{x^3}{3} - \frac{x^5}{5} - \dots, \quad |x| \leq 1$$

```python
import numpy as np

# 1. Задаємо параметри (згідно з умовою |x| <= 1)
x = 0.5
precision = 6  # кількість знаків після коми
epsilon = 10**(-precision)

# --- Спосіб 1: Математичні функції (ліва частина формули) ---
# Оскільки в numpy немає прямої функції arcctg, використовуємо тотожність:
# arcctg(x) = pi/2 - arctan(x)
res_math = np.pi / 2 - np.arctan(x)

# --- Спосіб 2: Сума ряду (права частина формули) ---
# Початкове значення з формули: pi/2
res_series = np.pi / 2
n = 0
term = 1  # ініціалізація для входу в цикл

while True:
    # Розрахунок n-го члена ряду за формулою: ((-1)^(n+1) * x^(2n+1)) / (2n+1)
    term = ((-1)**(n + 1) * (x**(2 * n + 1))) / (2 * n + 1)

    res_series += term

    # Перевірка точності: зупиняємося, коли модуль члена ряду менший за epsilon
    if abs(term) < epsilon:
        break

    n += 1

# Кількість підсумованих членів (n починається з 0, тому додаємо 1)
total_terms = n + 1

# 3. Вивід результатів
print(f"Параметри: x = {x}, точність = {precision} знаків")
print("-" * 40)
print(f"Значення через math функції:  {round(res_math, precision)}")
print(f"Значення через суму ряду:     {round(res_series, precision)}")
print(f"Кількість членів ряду:        {total_terms}")
```

## Завдання 3

```python
import numpy as np

# 1. Задаємо матрицю 10x10 вручну (без використання random)
# Створимо матрицю, де для демонстрації k=0 та k=5 будуть збігатися
matrix = np.array([
    [10,  1,  2,  3,  4,  5,  6,  7,  8,  9],  # k=0: рядок збігається зі стовпцем 0
    [ 1,  5,  0,  0,  0,  0,  0,  0,  0,  0],
    [ 2,  0,  8,  0,  0,  0,  0,  0,  0,  0],
    [ 3,  0,  0, 12,  0,  0,  0,  0,  0,  0],
    [ 4,  0,  0,  0,  9,  0,  0,  0,  0,  0],
    [ 5,  0,  0,  0,  0, 25,  0,  0,  0,  0],  # k=5: рядок збігається зі стовпцем 5
    [ 6,  0,  0,  0,  0,  0, 14,  0,  0,  0],
    [ 7,  0,  0,  0,  0,  0,  0, -2,  0,  0],  # Рядок з від'ємним елементом
    [ 8,  0,  0,  0,  0,  0,  0,  0, 18,  0],
    [ 9,  0,  0,  0,  0,  0,  0,  0,  0, -5]   # Ще один рядок з від'ємним елементом
])

# 2. Пошук k, де k-й рядок збігається з k-м стовпцем
matching_k = []
for k in range(10):
    # Порівнюємо k-й рядок та k-й стовпець
    if np.array_equal(matrix[k, :], matrix[:, k]):
        matching_k.append(k)

# 3. Знаходження суми елементів у рядках з хоча б одним від'ємним елементом
negative_rows_sum = 0
found_rows_indices = []

for i in range(10):
    # Перевірка, чи є у рядку хоча б один елемент < 0
    if np.any(matrix[i, :] < 0):
        current_row_sum = np.sum(matrix[i, :])
        negative_rows_sum += current_row_sum
        found_rows_indices.append(i)

# 4. Вивід результатів
print("Аналіз матриці 10x10:")
print("-" * 30)
print(f"Індекси k, де рядок збігається зі стовпцем: {matching_k}")
print(f"Індекси рядків з від'ємними елементами: {found_rows_indices}")
print(f"Загальна сума елементів у цих рядках: {negative_rows_sum}")
```
